```python
def make_hist(column, c):
    title = column.replace('_', ' ').title()
    
    figure = plt.figure(figsize=(12,6))
    plt.subplot(1, 2, 1)
    plt.hist(sync[sync['synced'] == 0][column], color = c)
    plt.title(f'Dist of Not Synced {title}')
    plt.xlabel(title)

    plt.subplot(1, 2, 2)
    plt.hist(sync[sync['synced'] == 1][column], color = c)
    plt.title(f'Dist of Synced {title}')
    plt.xlabel(title)
    
    return figure
```

```python
make_hist('s_dance', 'darkorange');
```
